```python
#División set de entrenamiento y set de pruebas
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.20, random_state = 0)
```

```python
from imblearn.over_sampling import SMOTE
#pip install imblearn

counter = Counter(y) # y es la variable objetivo (target) del dataset de entrenamiento
print('Antes',counter)
# oversampling el datased de entrenamiento usando SMOTE
smt = SMOTE()
#X_train, y_train = smt.fit_resample(X_train, y_train)
X_train, y_train = smt.fit_resample(X, y)
X_test, y_test = smt.fit_resample(X, y)

counter = Counter(y_train)
counter_test=Counter(y_test)
print('Después',counter)
print('Después test',counter_test)
```

```python
from sklearn.preprocessing import StandardScaler
sc_X = StandardScaler()
X_train = sc_X.fit_transform(X_train)
X_test = sc_X.transform(X_test)
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
classifier = LogisticRegression(random_state = 0)
classifier.fit(X_train, y_train)
```

```python
y_pred  = classifier.predict(X_test)
y_pred_train=classifier.predict(X_train)
```

```python
from sklearn.metrics import confusion_matrix
#Elaborar matriz de confusion
cm = confusion_matrix(y_test, y_pred)
```
